```python
import numpy as np
from scipy.stats import chi2_contingency

# Create the crosstab data as a NumPy array
data = np.array([
    [15, 20],  # Facebook
    [30, 35],  # Instagram
    [5, 15]    # TikTok
])

# Perform the Chi-Square test
chi2, p, dof, expected = chi2_contingency(data)

# Print the results
print(f"Chi-Square Statistic: {chi2}")
print(f"p-value: {p}")
print(f"Degrees of Freedom: {dof}")
print("Expected Frequencies:")
print(expected)
```

```python
import pandas as pd

# Create the raw data based on the crosstab
data = {
    'Gender': ['Male'] * 15 + ['Female'] * 20 +   # Facebook users
              ['Male'] * 30 + ['Female'] * 35 +   # Instagram users
              ['Male'] * 5 + ['Female'] * 15,     # TikTok users
    'Social Media': ['Facebook'] * 35 +           # 15 Males + 20 Females (Facebook)
                    ['Instagram'] * 65 +          # 30 Males + 35 Females (Instagram)
                    ['TikTok'] * 20               # 5 Males + 15 Females (TikTok)
}

# Create the DataFrame
df = pd.DataFrame(data)

# Show the first few rows of the original dataset
print(df.head())
```

```python
df.shape
```

```python
crosstab = pd.crosstab(df['Social Media'], df['Gender'])

# Perform the Chi-Square test on the crosstab
chi2, p, dof, expected = chi2_contingency(crosstab)

# Print the crosstab and the Chi-Square test results
print("Crosstab:")
print(crosstab)
print("\nChi-Square Statistic:", chi2)
print("p-value:", p)
print("Degrees of Freedom:", dof)
print("Expected Frequencies:")
print(expected)
```
